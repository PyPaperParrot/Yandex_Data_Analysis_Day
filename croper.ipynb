```python
from PIL import Image
```

```python
img = Image.open('blandwh.png')
count = 0
for x in range(1, 3602, 100):
    for y in range(1, 3602, 100):
        #if x == y:
        count+=1
        area = (x, y, x+98, y+98)
        cropped_img = img.crop(area)
        print(cropped_img.size)
        cropped_img.save('image_data/'+str(count)+'.png')
print('\n\n', count)
```
